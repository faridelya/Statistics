```python
import pandas as pd 
import seaborn as sn
import matplotlib.pyplot as plt
import io



pd.set_option('display.max_rows', 3000)
pd.set_option('display.max_columns', 37)
```

```python

cnty = pd.read_csv(io.BytesIO(uploaded['us_county_data.csv']))
cnty
```

# **Q1**
### **Generate a box plot showing the distribution of county-level median income(“med_income”) for all counties in the dataset.** 

```python
fig, ax = plt.subplots(figsize=(35,10))
b = sn.boxplot(data=cnty, x='med_income' , ax=ax);
b.axes.set_title("Box Plot",fontsize=50)
b.set_xlabel("median county",fontsize=20)
b.tick_params(labelsize=15)
plt.show()
```

# **Q2**
### **Generate a grouped box plot showing the distribution of county-level median income (“med_income”) grouped by subregion of the country (“SUB_REGION”) for all counties in the dataset.**

```python
fig, ax = plt.subplots(figsize=(35,10))
b = sn.boxplot(data=cnty, x='med_income' ,y='SUB_REGION' , hue="SUB_REGION",ax=ax, dodge=False);
b.axes.set_title("Box Plot",fontsize=50)
b.set_xlabel("median county income",fontsize=20)
b.set_ylabel("Sub Region",fontsize=20)
b.tick_params(labelsize=15)
plt.show()
```

# **Q3**
## **Generate a histogram showing the distribution of county-level median income(“med_income”) for all counties in the dataset**

```python
fig, ax = plt.subplots(figsize=(35,10))
b = sn.histplot(data=cnty,  x='med_income', y=None, hue=None, stat='count', bins='auto',  element='bars', ax=ax,);
b.axes.set_title("Box Plot",fontsize=50)
b.set_xlabel("median county income",fontsize=20)
b.set_ylabel("Count",fontsize=20)
b.tick_params(labelsize=15)
plt.show()
```

# **Q5**
## **Generate a histogram showing the distribution of county-level percent forest cover (“per_for”) for only counties occurring in the West North Central (“W N Cen”) subregion (“SUB_REGION”)**

```python
wn_cen = cnty[cnty['SUB_REGION'] == 'W N Cen']
fig, ax = plt.subplots(figsize=(35,10))
b = sn.histplot(data=wn_cen,  x='per_for', hue='SUB_REGION',kde=True,  ax=ax,);
b.axes.set_title("Histogram",fontsize=50)
b.set_xlabel("percent forest cover in  West North Central",fontsize=20)
b.set_ylabel("Count",fontsize=20)
b.tick_params(labelsize=15)
plt.show()
```

# **Q6**
## **Generate grouped kernel density plots showing the distribution of countylevel percent cropland (“per_crop”) for only counties occurring in the West North Central (“W N Cen”), Pacific (“Pacific”), and Mountain (“Mtn”) subregions (“SUB_REGION”). Each subregion should have a separate density plot in the graph  space**

```python
wn_cen =cnty[cnty['SUB_REGION'] == "W N Cen" ]
pf =cnty[ cnty['SUB_REGION'] ==  'Pacific']
mtn = cnty[cnty['SUB_REGION'] == 'Mtn']
wn_cen_pf = pd.concat([wn_cen,pf,mtn])

fig, ax = plt.subplots(figsize=(35,10))
b = sn.kdeplot(data=wn_cen_pf,  x='per_crop', hue='SUB_REGION',ax=ax);
b.axes.set_title("KDE",fontsize=50)
b.set_xlabel("PERCENT CROPLAND ",fontsize=20)
b.set_ylabel("Density",fontsize=20)
b.tick_params(labelsize=15)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(35,10))
sn.kdeplot(
   data=wn_cen_pf, x="per_crop", hue="SUB_REGION",
   fill=True, common_norm=False, palette="crest",
   alpha=.5, linewidth=0,ax=ax
);
```

# **Q8**


```python
 # 1st standar deviation 68% distribution from mean
import numpy as np
 
wn_cen =cnty[cnty['SUB_REGION'] == "W N Cen" ]
pf =cnty[ cnty['SUB_REGION'] ==  'Pacific']
mtn = cnty[cnty['SUB_REGION'] == 'Mtn']
wn_cen_pf = pd.concat([wn_cen,pf,mtn])
fig, ax = plt.subplots(figsize=(20,10))
 b =sn.barplot(x="SUB_REGION", y='per_crop', data=wn_cen_pf, estimator=np.mean, ci=68, capsize=.68, color='lightblue')
b.axes.set_title("adding error bar to mean percent per crop land",fontsize=30)
b.set_xlabel("Region ",fontsize=20)
b.set_ylabel("Percentage",fontsize=20)
b.tick_params(labelsize=15)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(30,13))
sn.pointplot(data=wn_cen_pf, x='SUB_REGION', y='per_crop', hue='SUB_REGION', errorbar=("sd",68 ),ax=ax);
```

# **Q9**
## Create a scatter plot showing the relationship between mean elevation(“dem”) and mean county-level average annual temperature (“tempmn”) for all counties in the country. 

```python
fig, ax = plt.subplots(figsize=(38,12))
b = sn.scatterplot(
    data=cnty, x="dem", y="tempmn", 
    s=200, );

b.axes.set_title("Scatter plot relationship btw dem and tempmn",fontsize=30)
b.set_xlabel("dem ",fontsize=20)
b.set_ylabel("tempmn",fontsize=20)
b.tick_params(labelsize=15)
plt.show()
```
